```python
import pickle
file_m = pickle.load(open('file_m_40_hc2_LAL.pkl', 'rb'))
file_e= pickle.load(open('file_e_1.1_hc2_LAL.pkl', 'rb'))
file_b= pickle.load(open('file_b_60_hc2_LAL.pkl', 'rb'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
```

```python
arrlen=np.shape(file_m)[-1]
arrlen
```

```python
matchfactor=0.99;
```

```python
eref1=np.array([1.1,1.15,1.2,1.25])
bref1=np.array([50,60,80,100]) 

ecc1=np.linspace(1.09,1.4,arrlen)
impacts1=np.linspace(40,150,arrlen)
```

# CONS M

```python
eref1=np.array([1.1,1.15,1.2,1.25])
bref1=np.array([50,60,80,100]) 

ecc1=np.linspace(1.09,1.4,arrlen)
impacts1=np.linspace(40,150,arrlen)
#######################################
s=1
fig=plt.figure(figsize=(10,7))
for i in range(len(eref1)):
    for j in range(len(bref1)):
        plt.subplot(4,4,s)
        #c1=next(color)
        ind=np.where(file_m[i,j]>=matchfactor)
        plt.scatter(ecc1[ind[0]],impacts1[ind[1]],linewidths=2,alpha=0.05)
        #plt.contourf(ecc1, impacts1, file_m[i,j],[matchfactor,1],alpha=0.1)
        plt.scatter(eref1[i],bref1[j],color='k')
        plt.title('e='+str(eref1[i])+', b='+str(bref1[j]))
        s+=1
plt.suptitle('M=40; matchfactor='+str(matchfactor))   
# plt.xlim([1.09,1.4])
# plt.ylim([2.8,150])
fig.tight_layout()     
plt.show()
```

# CONS e

```python
mref1=np.array([20,40,60,100])
bref1=np.array([50,60,80,100]) 

masses1=np.linspace(10,150,arrlen)
impacts1=np.linspace(40,150,arrlen)
###########################################
s=1
fig=plt.figure(figsize=(10,7))
for i in range(len(mref1)):
    for j in range(len(bref1)):
        plt.subplot(4,4,s)
        #c1=next(color)
        ind=np.where(file_e[i,j]>=matchfactor)
        plt.scatter(masses1[ind[0]],impacts1[ind[1]],linewidths=2,color='k',alpha=0.05)
        #plt.contourf(ecc1, impacts1, file_m[i,j],[matchfactor,1],alpha=0.1)
        plt.scatter(mref1[i],bref1[j],color='g')
        plt.title('M='+str(mref1[i])+', b='+str(bref1[j]))
        s+=1
plt.suptitle('e=1.1; matchfactor='+str(matchfactor)) 
fig.tight_layout()       
```

# CONS b

```python
mref1=np.array([20,40,60,100])
eref1=np.array([1.1,1.15,1.2,1.25])

masses1=np.linspace(10,150,arrlen)
ecc1=np.linspace(1.09,1.4,arrlen)
####################################################################
s=1
fig=plt.figure(figsize=(10,7))
for i in range(len(mref1)):
    for j in range(len(eref1)):
        plt.subplot(4,4,s)
        #c1=next(color)
        ind=np.where(file_b[i,j]>=matchfactor)
        plt.scatter(masses1[ind[0]],ecc1[ind[1]],linewidths=2,alpha=0.05,color='salmon')
        #plt.contourf(ecc1, impacts1, file_m[i,j],[matchfactor,1],alpha=0.1)
        plt.scatter(mref1[i],eref1[j],color='purple')
        plt.title('M='+str(mref1[i])+', e='+str(eref1[j]))
        s+=1
        plt.xlim([10,150])
        plt.ylim([1.09,1.4])
plt.suptitle('b=60; matchfactor='+str(matchfactor))

fig.tight_layout()   
```
